9: Q.9. Write a code to create the given structure using only for loop.
*
***
*****
*******
*********

```python
def create_structure(rows):
    for i in range(1, rows * 2, 2):  # Loop for the odd number of stars in each row
        print('*' * i)

# Define the number of rows for the structure
num_rows = 5

# Call the function to create the structure with the specified number of rows
create_structure(num_rows)
```

10: Write a code to create the given structure using while loop.
|||||||||
|||||||
|||||
|||
|

```python
# ans: 10
def create_structure(rows):
    length = rows * 2 - 1  # Calculate the initial length of pipes

    while length >= 1:
        print('|' * length)  # Print the row with pipes
        length -= 2  # Decrement the length by 2 for the next row

# Define the number of rows for the structure
num_rows = 5

create_structure(num_rows)
```
