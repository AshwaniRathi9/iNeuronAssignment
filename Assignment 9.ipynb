One difference between lambda and regular functions is that lambda functions are nameless, while regular functions have a name.

5: Are lambda functions in Python able to access variables defined outside of their own scope? Explain with an example.

:- Yes, lambda functions in Python can access variables defined outside of their own scope. This is known as “closing over” a variable. When a lambda function is defined, it captures the current value of any variables that are used within its body.

For example, the lambda in button.clicked.connect (lambda: self.squareButtonHandler (buttonNumber)) is an inner function that contains a reference to buttonNumber in the outer scope.

7: Create a lambda function to find the maximum value in a list of integers.

```python
from functools import reduce
lst = [22, 36, 47, 2, 13]
reduce(lambda x, y: x if (x > y) else y, lst)
```
